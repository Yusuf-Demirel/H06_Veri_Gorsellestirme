```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/workspaces/H06_Veri_Gorsellestirme/datalar/flight_delays (1).csv')
df
```

```python
plt.figure(figsize=(10,6))
plt.title('Aylara Göre Uçuşların Ortalama Gecikm Süresi Dakika Cinsinden')
sns.barplot(x=df.index, y=df['NK'])
```

# HEATMAPS

Heatmap bütün sütunların sürekli sayısal verilerini kıyaslamak için kullanabiliriz.

```python
plt.figure(figsize = (14,7))
plt.title('Aylara Göre Her Havalimanının Uçuş Gecikmelerini Dakika Cinsinden')
sns.heatmap(data= df, annot= True)

plt.xlabel('Havalimaları')
plt.ylabel('Aylar')
plt.show() # Görselin kendisini döndürmesi için lazımdır.
```

```python
df = pd.read_csv('/workspaces/H06_Veri_Gorsellestirme/datalar/SampleSuperstore (1).csv')
df
```

```python
num_variable = ['Sales','Quantity','Discount','Profit']
```

```python
fig, ax = plt.subplots(1,4,figsize = (10,6))
plt.subplots_adjust(wspace = 0.5)
colors = ['brown','g','y','None']
for i,variable in enumerate(num_variable):
    sns.boxplot(data = df[variable],ax = ax[i], color = colors[i])
    ax[i].set_xlabel(num_variable[i])
    ax[i].set_xticklabels([])

plt.show()
```
